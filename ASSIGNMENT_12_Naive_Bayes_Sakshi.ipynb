# Naive Bayes

```python
x_train = data1.iloc[:,0:13]
y_train = data1.iloc[:,13]
x_test = data2.iloc[:,0:13]
y_test = data2.iloc[:,13]
```

```python
from sklearn.naive_bayes import MultinomialNB as MB
from sklearn.naive_bayes import GaussianNB as GB
```

## MultinomialNB

```python
classifier_mb = MB()
```

```python
classifier_mb.fit(x_train, y_train)
```

```python
y_pred_mnb = classifier_mb.predict(x_train)
```

```python
y_pred_mnb
```

```python
y_pred_mnb.shape
```

```python
y_train.shape
```

```python
accuracy_train_m = np.mean(y_pred_mnb==y_train)
```

```python
accuracy_train_m 
```

```python
np.mean(y_pred_mnb==y_train.values)
```

```python
print('Accuracy of Training MultinomialNB =',np.mean(y_pred_mnb==y_train.values),'\n','Accuracy of testing MultinomialNB =',accuracy_train_m)
```

## GaussianNB

```python
classifier_gb = GB()
```

```python
classifier_gb.fit(x_train, y_train)
```

```python
y_pred_gnb = classifier_gb.predict(x_test)
```

```python
y_pred_gnb.shape
```

```python
y_test.shape
```

```python
np.mean(y_pred_gnb==y_test.values)
```

```python
accuracy_test_g = np.mean(y_pred_gnb==y_test)
```

```python
accuracy_test_g
```

```python
print('Accuracy of Training GaussianNB=',np.mean(y_pred_gnb==y_test.values),'\n',
      
      'Accuracy of Training GaussianNB=',accuracy_test_g)
```

## Conclusion 

## **GaussianNB Model has a better Accuracy**
